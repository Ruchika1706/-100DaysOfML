Python code to implement train_test_split function 

```python
import numpy as np
def train_test_split(data1, data2, test_ratio, seed):
    #print(data1)
    #print(data2)
    #print(data1.size) gives 214 * 10 -> 2140
    #print(data2.size) gives 214 * 1 -> 214
    np.random.seed(seed)
    test_size = int(len(data1.index) *(test_ratio))
    x_test_indices = data1[:test_size]
    x_train_indices = data1[test_size:]
    test_size = (int(data2.size * test_ratio))
    y_test_indices = data2[:test_size]
    y_train_indices = data2[test_size:]
    return x_train_indices, x_test_indices, y_train_indices, y_test_indices

import pandas as pd
columns = ["ID", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type"]
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data", header=None, names=columns)
df1 = df[["ID", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe"]]
df2 = df[["Type"]]
X_train, X_test, y_train, y_test = train_test_split(df1, df2, 0.2, 42)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```
