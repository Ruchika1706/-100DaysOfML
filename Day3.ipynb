```python
#Series Example 1
from pandas import Series, DataFrame
s1 = Series(range(0,4))
s2 = Series(['a','b'])*3
print(s1.dtype)
print(s1.index) #as Series is one dimensional so it is for columns O/P RangeIndex(start=0, stop=4, step=1)
print(s2)
```

```python
#DataFrame Example 1
from pandas import DataFrame
df = DataFrame(data=[1,2,3,4])
print(df.columns) #O/P RangeIndex(start=0, stop=1, step=1) as only 1 column, provides axes labels for column 
print(df.index) #O/P RangeIndex(start=0, stop=4, step=1) as 4 rows, provides axes labels for rows 
print(df.index.tolist()) #to convert the index to a list.
print(df)

columns = ['x','y','z']
df2 = DataFrame(data={'a':[123],'b':[456],'c':[345]})
df2.columns = columns
print(df2)

#Index functions
idx = df2.columns
l = idx.tolist() # get as python list
#a = idx.values() # get as numpy array
label = idx.min() #minimum label
label2 = idx.max()#maximum label
print(l)
#print(a)
print(label)
print(label2)
```

```python
#Data in Series combine into a DataFrame
import pandas as pd
from pandas import DataFrame, Series
s1 = Series(range(6))
print(s1) # [0,1,2,3,4,5]
s2 = s1*s1
print(s2) # [0,1,4,9,16,25]
s2.index = s2.index + 2 #misalign indexes, s2 index now starts from 2,3,4,5,6,7
print(s2)
df = pd.concat([s1,s2],axis = 1)
print(df)

s3 = Series({'Tom':1, 'Dick':2,'Harry':3})
s4 = Series({'Tom':4, 'Dick':5,'Harry':6})
print(s3)
print(s4)
print(s3+s4)
df = pd.concat({'A':s3,'B':s4}, axis=1) #axis is the axis along which to combine the data 0 for index, 1 for columns, as pe
#my understanding, treat the other axes as a union of all possible values from the two series
print(df)

df = pd.concat({'A':s3,'B':s4}) #default axis is 0
print(df)
```

```python
#Get Data into DataFrame from a Dictionary
from pandas import DataFrame
import pandas as pd
df = DataFrame({'col0':{
    'row0':[1],
    'row1':[3]
    },
    'col1':{
    'row0':[2],
    'row1':[4]},
})
print(df)
print(df.size) # row count * column count
print(df.dtypes) #Series column data types
(r,c) = df.axes
print(r,c)

idx = df.columns
print(idx) #print column indexes

label = df.columns[0] # get first column label
print(label)

lst = df.columns.tolist() # get as a list
print(lst)

s = df['col0'] # select col to series
print(s)
print(type(s)) #<class 'pandas.core.series.Series'>

#to swap column contents
df[['col0','col1']] = df[['col1','col0']]
print(df)

print(df)
df = df[['col0']] #Select col to df
print(df)
print(type(df)) #<class 'pandas.core.frame.DataFrame'>

```
