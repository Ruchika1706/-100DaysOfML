```python
import pandas as pd
import numpy as np
columns = ["ID", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type"]
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data", header=None, names=columns)
print(df)
```

```python
df.head()
df.info()
df.describe()
```

As per the current results, no data is missing in the given data set. 214 rows * 11 columns

```python
df['Type'].value_counts()
```

df['Type'] is a categorical attribute. df['Type'] in my understanding is the Y i.e. the Label, all other columns are the X i.e attributes. Let us divide the dataframe into X and Ys

```python
df1 = df[["ID", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe"]]
df2 = df[["Type"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1, df2, test_size=0.30, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
clf = GaussianNB()
clf.fit(X_train, y_train)
y = clf.predict(X_test)
accuracy_score(y_test, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1, df2, test_size=0.20, random_state=43)
clf.fit(X_train, y_train)
y = clf.predict(X_test)
accuracy_score(y_test, y)
```
